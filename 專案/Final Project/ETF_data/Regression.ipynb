```r
# Price = α+β1(Score) +μ 
regg_SPY <- lm(SPY$Price ~ SPY$Score)
summary(regg_SPY)
```

從以上的分析表得出β1接受虛無假設：β1 = 0，故SPY(Equity)的價格變動和情緒分數變動無顯著關係
